```python
import math
from tensorflow.keras.constraints import Constraint
from tensorflow.keras import initializers

# Custom constraint for transpose matrix
class TransposeConstraint(Constraint):
    def __call__(self, w):
        return tf.transpose(w)

def gauss(x, m, var=1.):  #var=0.1
    coeff = tf.math.rsqrt(2*math.pi*var)
    exp = tf.math.exp(-tf.reduce_sum((x - m)**2, axis=-1, keepdims=False)/(2*var))
    return coeff*exp #, coeff, exp


class Encoder(Model):
    def __init__(self, input_dim, hidden_dim):
        super(Encoder, self).__init__()
        self.dense = layers.Dense(hidden_dim, activation=None,kernel_initializer=initializers.Zeros(),use_bias=False) # 'relu'

    def call(self, x):
        h = self.dense(x)
        z = tf.nn.softmax(h)
        return h, z  # shape=(Batch, hidden_dim)

class Sampler(Model):
    def __init__(self, n_sample, hidden_dim):
        super(Sampler, self).__init__()
        self.n_sample = n_sample
        self.hidden_dim = hidden_dim

    def call(self, z):
        p = z
        sample = tf.reshape(tf.random.categorical(tf.math.log(p), num_samples=self.n_sample), [self.n_sample, z.shape[0]])
        onehot = tf.one_hot(sample, depth=self.hidden_dim)
        #print(z[0],sample[0],)
        return onehot  # shape=(n_sample, Batch, hidden_dim)

class Decoder(Model):
    def __init__(self, input_dim, hidden_dim):
        super(Decoder, self).__init__()
        self.dense_output = layers.Dense(input_dim, activation=None,kernel_initializer=initializers.Zeros(),use_bias=False) # 'sigmoid'

    def call(self, onehot):
        output = self.dense_output(onehot)
        return output  # shape=(n_sample, Batch, input_dim)

class SAE(Model):
    def __init__(self, input_dim, hidden_dim, n_sample):
        super(SAE, self).__init__()
        self.encoder = Encoder(input_dim, hidden_dim)
        #my_tensor = tf.zeros(shape=[input_dim,hidden_dim])
        #my_variable = tf.Variable(my_tensor,trainable=True)
        #print(self.trainable_variables)
        #print(self.trainable_variables.append(my_variable))
        #print(self.trainable_variables)
        self.sampler = Sampler(n_sample,  hidden_dim)
        #self.decoder = Decoder(input_dim, hidden_dim)
        #self.decoder.set_weights(self.encoder.weights[0].T)
        #self.sampler = self.encoder.weights[0]

    def call(self, x):
        h, z   = self.encoder(x)
        #h = tf.tensordot(x, self.trainable_variables, axes=[2, 2])
        #z = tf.nn.softmax(h)
        onehot = self.sampler(z)
        #x_decoded = self.decoder(onehot)
        x_decoded = tf.tensordot(onehot, self.encoder.weights[0], axes=[2, 1])
        #x_decoded = tf.tensordot(onehot,self.trainable_variables, axes=[2, 1])

        temp = gauss(tf.expand_dims(x, axis=0), x_decoded)  # shape=(n_sample, Batch)
        weight = temp / tf.reduce_mean(temp, axis=0)
        #weight = tf.Variable(weight, trainable=False)
        #print(tf.expand_dims(x, axis=0))

        return h, z, onehot, x_decoded, weight
```

```python
image_size = x_train.shape[-1]
h_dim = 100
n_sample = 5 #10

slse = SAE(input_dim=image_size,
           hidden_dim=h_dim,
           n_sample=n_sample)
slse.build(input_shape=x_train.shape)
slse.summary()


#print(slse.trainable_variables)

optimizer = tf.keras.optimizers.Adam(beta_1=0.9, beta_2=0.98, epsilon=1e-9)
```

```python
n_patterns = 500
x_patterns = x_train[:n_patterns]
for i in range(n_patterns):
    x_patterns[i] = 10*x_patterns[i]/np.linalg.norm(x_patterns[i])

x_memory = x_train[n_patterns:(n_patterns+100)]
for j in range(50):
    x_memory[j] = 10*x_memory[j]/np.linalg.norm(x_memory[j])

np.set_printoptions(threshold=100)
#print(slse.trainable_variables)
#print(x_patterns[:20].T)
slse.encoder.set_weights([x_memory.T])
#print(slse.trainable_variables)

#BUFFER_SIZE = 256
BATCH_SIZE = 50 # n_patterns #256
dataset = tf.data.Dataset.from_tensor_slices((x_patterns))#.shuffle(BUFFER_SIZE)
dataset = dataset.batch(BATCH_SIZE, drop_remainder=True)
#steps_per_epoch = len(x_train) // BATCH_SIZE
```

```python
def loss_function(x, predict, z, onehot, weight):
    mse = losses.MSE(x, predict)  # shape=(n_sample, Batch)
    eps = 1e-12
    logsoftmax = tf.math.log(tf.reduce_sum(z*onehot, axis=-1) + eps)  # shape=(n_sample, Batch)
    loss = tf.reduce_mean(weight * (mse - logsoftmax))  # mean through n_sample & Batch
    if np.isnan(loss):
        #print(weight,np.linalg.norm(weight),weight.shape)
        loss = 0
    return loss

#@tf.function
def train_step(x):
    loss = 0
    with tf.GradientTape() as tape:
        h, z, onehot, x_reconstructed, weight = slse(x, training=True)
        weight = tf.Variable(weight, trainable=False)
        loss += loss_function(x, x_reconstructed, z, onehot, weight)
    batch_loss = (loss / len(x))
    variables = slse.trainable_variables
    gradients = tape.gradient(loss, variables)
    optimizer.apply_gradients(zip(gradients, variables))

    return batch_loss
```

```python
EPOCHS = 1001 #3200 #2400 #1600 #800 #200 # 300

train_loss = []
for epoch in range(EPOCHS):
    for batch, x in enumerate(dataset):
        batch_loss = train_step(x)

    train_loss.append(batch_loss)
    if batch_loss>0.1:
        weight_er = slse.layers[0].weights[0].numpy().T
    if (epoch+1)%20==1:
        print(f'Epoch {epoch+1}, Batch {batch}, Train Loss {train_loss[epoch]}')  # Loss {}, batch_loss.numpy()
```

```python
class RLSEGate(layers.Layer):
    def __init__(self, units_hidden, **kwargs):
        super().__init__(**kwargs)
        self.units_hidden = units_hidden
        self.threshold = tf.math.log(tf.constant([units_hidden], dtype=float))  #  e^{0.1, 0.5, 0.8, 1.} = 1.105, 1.648, 2.225, 2.718
        print(f"threshold = {self.threshold}")

    def get_config(self):
        config = super().get_config()#.copy()
        config.update({
            "units_hidden": self.units_hidden,
        })
        return config

    #def build(self, input_shape):
    #    self.dim = input_shape[-1]
    #
    #    self.w1 = self.add_weight(shape=(self.dim, self.units_hidden),
    #                              initializer='glorot_uniform',
    #                              trainable=True,
    #                              name='xi')

    def call(self, x):
        logsumexp_x = tf.math.reduce_logsumexp(x, axis=-1, keepdims=True)
        mask = tf.experimental.numpy.heaviside(logsumexp_x - self.threshold, 0)
        return  mask*tf.nn.softmax(x), mask,logsumexp_x  #, tf.squeeze(logsumexp_x)


def OTA_MHN(input):
    h_dim = 20000
    RLSElayer = RLSEGate(h_dim)
    h, z = slse.layers[0](input)
    x, mask,lse = RLSElayer(h)
    x=mask*tf.tensordot(z, slse.layers[0].weights[0], axes=[1, 1])
    return x,lse
```

```python
def smooth_curve(points, factor=0.9):
    smoothed_points = []
    for point in points:
        if len(smoothed_points)==0:
            smoothed_points.append(point)
        else:
            previous = smoothed_points[-1]
            smoothed_points.append(previous*factor + point*(1-factor))
    return smoothed_points

smoothed_loss = smooth_curve(train_loss)

plt.plot(smoothed_loss)
#plt.plot(train_loss)
plt.title('Train Loss')
plt.xlabel("epoch")
#plt.ylim((0, z.numpy().max()))
plt.show()
```
